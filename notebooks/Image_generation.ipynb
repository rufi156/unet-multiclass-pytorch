# Generate images

```python
import sys
sys.path.insert(0, "..")
import matplotlib.pyplot as plt

from lib import make_image, make_image_dir, save_images, mask2rgb
```

### Create training images

```python
width,height = 200,400
n_images = 10
export_dir = '../data/images_test'

images = [make_image(width,height) for i in range(n_images)]
save_images(export_dir, images)
```

### Image and mask sample

```python
img = images[0][0]
mask = images[0][1]

fig, ax = plt.subplots(1, 2, figsize=(10, 8))
ax[0].imshow(img, cmap='gray')
ax[0].axis('off')
ax[1].imshow(mask2rgb(mask))
ax[1].axis('off')
```
